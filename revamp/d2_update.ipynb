```python
from dotenv import load_dotenv
from spotipy.oauth2 import SpotifyClientCredentials
import spotipy
import pandas as pd 
import sqlite3
from sqlite3 import Error

# Load environment variables 
load_dotenv()
sp = spotipy.Spotify(client_credentials_manager=SpotifyClientCredentials())
```

```python
fav_artists = [
    "Drake",
    "Becky G",
    "Jack Harlow",
    "Karol G",
    "Bad Bunny",
    "J Balvin",
    "Grupo Firme",
    "Queen",
    "Lynyrd Skynyrd",
    "Nirvana",
    "Led Zeppelin",
    "Billie Eilish",
    "J. Cole",
    "Rihanna",
    "Foo Fighters",
    "Natti Natasha",
    "Black Sabbath",
    "Van Halen",
    "Pink Floyd",
    "Britney Spears"
]
```

```python
def check_if_valid_data(df: pd.DataFrame) -> bool:
    if df.empty:
        print("No data available")
        return False 
    
    if df.isnull().values.any():
        num_nulls = df.isnull().sum().sum()
        raise Exception(f'{num_nulls} null values found')
    
    return True
```

<h1>ARTISTS</h1>

```python
artist_response = []

for artist in fav_artists:
        search_artist = sp.search(artist, limit=1, type="artist")

        artist_response.append(search_artist)

artist_df = pd.DataFrame.from_dict(artist_response)
artist_df
```

```python
artists_list = artist_df['artists'].tolist()
artists_list
```

<h3>That was a mess to go through so I thought it would be easier to get each artist's info one at a time and add it to the dataframe since there were only 20 artists and the layout of the info was consistent.</h3>

```python
artist_response = []

for artist in fav_artists:
        search_artist = sp.search(artist, limit=1, type="artist")
        access_artist_info = search_artist["artists"]["items"][0]

        # Save the data to dictionary
        artist_info = {
            "artist_id": access_artist_info["id"],
            "artist_name": access_artist_info["name"],
            "external_url": access_artist_info["external_urls"]["spotify"],
            "genre": access_artist_info["genres"][0],
            "image_url": access_artist_info["images"][0]["url"],
            "followers": access_artist_info["followers"]["total"],
            "popularity": access_artist_info["popularity"],
            "type": access_artist_info["type"],
            "artist_uri": access_artist_info["uri"]
        }

        artist_response.append(artist_info)

artist_df = pd.DataFrame.from_dict(artist_response)
artist_df
```

```python
# remove duplicates 
artist_df.drop_duplicates(subset="artist_id")
```

```python
artist_df.info()
```

<h3>All of our data types are correct and no null values.</h3>


```python
check_if_valid_data(artist_df)
```

<h1>ALBUMS</h1>

<h3>I would get all the artist ids from the pandas dataframe</h3>

```python
artist_ids = artist_df['artist_id'].tolist()
artist_ids
```

```python
album_response = []

# Looping through all the artist_ids
for artist_id in artist_ids:

    # Pulling data via spotipy with an increased limit to get all the artists albums
    albums = sp.artist_albums(artist_id, album_type="album", limit=50, country="US")

    # Looping through all the albums available 
    for album in albums["items"]:
        # Inserting the data into a dictionary
        album_info = {
            "album_id": album["id"],
            "album_name": album["name"],
            "external_url": album["external_urls"]["spotify"],
            "image_url": album["images"][0]["url"],
            "release_date": album["release_date"],
            "total_tracks": album["total_tracks"],
            "type": album["type"],
            "album_uri": album["uri"],
            "artist_id": album["artists"][0]["id"],
        }
        album_response.append(album_info)


album_df = pd.DataFrame.from_dict(album_response)
album_df

```

```python
# remove duplicates 

album_df.drop_duplicates(subset=["album_name", "artist_id"], inplace=True)
album_df
```

<h3>It removed 43 duplicate albums. Now I am going to remove deluxe albums.</h3>

```python
album_df = album_df[album_df["album_name"].str.contains("(Deluxe|Deluxe)|Deluxe Edition|Deluxe Version|Super Deluxe") == False]

album_df
```

<h1>TRACKS</h1>

```python
album_ids = album_df['album_id'].tolist()
album_ids
```

```python
track_response = []

# Going through each album id and searching for it's tracks
for album_id in album_ids:
    tracks = sp.album_tracks(album_id)

    # Adding data to dictionary 
    for track in tracks["items"]:
        track_info = {
            "track_id": track["id"],
            "song_name": track["name"],
            "external_url": track["external_urls"]["spotify"],
            "duration_ms": track["duration_ms"],
            "explicit": track["explicit"],
            "disc_number": track["disc_number"],
            "type": track["type"],
            "song_uri": track["uri"],
            "album_id": album_id,
        }

        track_response.append(track_info)

track_df = pd.DataFrame.from_dict(track_response)
track_df
```

```python
track_df.drop_duplicates(subset=["song_name", "album_id"], inplace=True)
track_df
```

```python
check_if_valid_data(track_df)
```

<h1>TRACK FEATURES</h1>

```python
track_ids = track_df["track_id"].tolist()
track_ids
```

```python
num_of_tracks = len(track_ids)

starting_idx = 0
stopping_idx = 100

feature_response = []

# Loop to make API calls with a maximum of 100 tracks at a time 
for i in range((num_of_tracks // 100) + 1):
    features = sp.audio_features(track_ids[starting_idx:stopping_idx])

    # Add data to the dictionary 
    for feature in features:
        feature_info = {
            "track_id": feature["id"],
            "danceability": feature["danceability"],
            "energy": feature["energy"],
            "instrumentalness": feature["instrumentalness"],
            "liveness": feature["liveness"],
            "loudness": feature["loudness"],
            "speechiness": feature["speechiness"],
            "tempo": feature["tempo"],
            "type": feature["type"],
            "valence": feature["valence"],
            "song_uri": feature["uri"],
        }

        feature_response.append(feature_info)
    
    # Increase our indexes and move the window up 
    starting_idx = stopping_idx
    stopping_idx += 100

feature_df = pd.DataFrame.from_dict(feature_response)
feature_df
```

```python
feature_df.drop_duplicates(subset="track_id", inplace=True)
feature_df
```

<h3>No duplicates and there are the same amount of tracks as features.</h3>

```python
feature_df.info()
```

<h3>All of our data types are correct and no null values.</h3>

```python
check_if_valid_data(feature_df)
```
